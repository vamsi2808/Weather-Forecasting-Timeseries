```python
# Adding a new Coloum in the table as season
import pandas as pd
import datetime as dt

df = pd.DataFrame({'date': pd.date_range('2000-01-01', '2001-01-01', periods=12)})
seasons = {(11, 12, 1, 2): 'winter', (3, 4, 5, 6, 7): 'summer', (8, 9, 10): 'rainy'}
df['month'] = df.date.dt.month

def season(ser):
    for k in seasons.keys():
        if ser in k:
            return seasons[k]

df['season'] = df.month.apply(season)
```

```python
df
```

Seasonal Difference with Mean

```python
gb=df.groupby(['date','season'])['month'].mean().reset_index()
plt.figure(figsize=(15,8))
sns.barplot(x='season',y='month',data=gb);
```
